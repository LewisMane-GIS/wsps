```python
import pandas as pd
import geopandas as gpd
import folium
from folium import plugins

# loading layers first
land = gpd.read_file("./data/land-in-chuka.geojson")
manholes = gpd.read_file("./data/manholes.geojson")
mt = gpd.read_file("./data/meru-tharaka.geojson")

sewer = gpd.read_file("./data/sewer-extent.geojson")
sewer_lines = gpd.read_file("./data/sewer-extent.geojson")
trunks = gpd.read_file("./data/trunk-sewer.geojson")
manlines = gpd.read_file("./data/manlines.geojson")
counties = gpd.read_file("./data/Counties.geojson")
                         
# first look
m = folium.Map(location=[-0.31894676971625324, 37.65712276600076], zoom_start=14, control_scale=True)


# first try for the layers
Manholes = folium.GeoJson(manholes, name='Manholes',
                         style_function = lambda x: {
                           'color': 'black'
                       },control=True).add_to(m)

Manlines = folium.GeoJson(manlines, name='Proposed Sewer Lines',
                         style_function = lambda x: {
                           'color': 'brown',
                           'weight': 2
                       },control=True).add_to(m)

lines = folium.GeoJson(sewer_lines, name='Sewerage').add_to(m)

Trunks = folium.GeoJson(trunks, name='Trunks Sewer',
                       style_function = lambda x: {
                           'color': 'green',
                           'weight': 2
                       },control=True).add_to(m)

Land = folium.GeoJson(land, name='Land').add_to(m)
Meth = folium.GeoJson(mt, name='Meru-Tharaka area').add_to(m)
Sewer = folium.GeoJson(sewer, name='Sewer Extent').add_to(m)
Kenya = Kenya = folium.GeoJson(counties, name='Counties',
                       style_function = lambda x: {
                           'color': 'black',
                           'weight': 1,
                           "opacity": 1,
                           'fillOpacity': 0,
                           'interactive': False
                       },
                       control=False).add_to(m)
                         
# adding basemap layers
folium.raster_layers.TileLayer('Open Street Map').add_to(m)
folium.raster_layers.TileLayer('Stamen Terrain').add_to(m)
folium.raster_layers.TileLayer('Stamen Toner').add_to(m)
folium.raster_layers.TileLayer('Stamen Watercolor').add_to(m)
folium.raster_layers.TileLayer('Cartodb Positron').add_to(m)
folium.raster_layers.TileLayer('Cartodb Dark_Matter').add_to(m)


# adding the layer control
folium.LayerControl().add_to(m)

# display

m.save("C:/Users/LEWIS/Desktop/TWWDA/New app 2/maps/chuka.html")
                         
                         
```
